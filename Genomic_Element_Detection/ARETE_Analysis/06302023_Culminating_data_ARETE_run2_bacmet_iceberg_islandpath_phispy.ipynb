```python
#Import mobsuite results from all the HLF genomes and combine into one big table.
import pandas as pd
from pathlib import Path
#create an empty dataframe
all_BACMET_results = []
#Determine the path that will lead to your mobsuite results files for all HLF genomes
BACMET_results_directory = "/media/haley/My Book1/Ecoli_results/Ecoli_ARETE_Results_Organized/VIDO_ARETE/bacmet/"
for BACMET_output in Path(BACMET_results_directory).glob('EC_*/*_BACMET.txt'):
    
    isolate_name = BACMET_output.parts[8]

    print(f"Parsing {BACMET_output}")
#Read tables in mobsuite output files with new header we designate
    Isolate_BACMET_results = pd.read_table(BACMET_output, 
                                       sep='\t',  
                                       names=['qseqid', 'sseqid', 'pident', 'slen', 'qlen', 'length', 'mismatch', 'gapopen', 'qstart', 'qend', 'sstart', 'send', 'evalue', 'bitscore', 'sequence'], 
                                            usecols=['qseqid', 'sseqid', 'pident', 'slen', 'qlen', 'length', 'mismatch', 'gapopen', 'qstart', 'qend', 'sstart', 'send', 'evalue', 'bitscore', 'sequence'])
#Put Plasmid_name (the name of the isolate) into a column in the table
    Isolate_BACMET_results['Isolate_Pathname'] = isolate_name
#Append the table to your empty dataframe
    all_BACMET_results.append(Isolate_BACMET_results)
#concatentate all the tables together
all_BACMET_results = pd.concat(all_BACMET_results)
#Print to see new table
print(all_BACMET_results)
```

```python
BACMET_reports_indexed=all_BACMET_results.set_index('Isolate_Pathname')
```

```python
BACMET_reports_indexed.to_csv('06302023_EcoliARETE_BACMET_unfiltered.csv')
```

```python
#set cutoff of 80% ID
BACMET_80_cutoff = BACMET_reports_indexed.loc[BACMET_reports_indexed['pident'] > 80]
BACMET_80_cutoff
```

```python
BACMET_80_cutoff.to_csv('06302023_EcoliARETE_BACMET_filtered80.csv')
```

```python
BACMET_80_cutoff['Gene_Name'] = BACMET_80_cutoff['sseqid'].str.split('|').str[1]
```

```python
BACMET_80_cutoff.to_csv("06302023_EcoliARETE_BACMET_filtered80_gene_name.csv")
```

```python
#Import islandpath results from all the HLF genomes and combine into one big table.
import pandas as pd
from pathlib import Path
#create an empty dataframe
all_islandpath_results = []
#Determine the path that will lead to your islandpath results files for all HLF genomes
islandpath_results_directory = "/media/haley/My Book1/Ecoli_results/Ecoli_ARETE_Results_Organized/VIDO_ARETE/islandpath"
for islandpath_output in Path(islandpath_results_directory).glob('EC_*/EC_*.gff'):
    
    isolate_name = islandpath_output.parts[8]

    print(f"Parsing {islandpath_output}")
#Read tables in islandpath output files with new header we designate
    Isolate_islandpath_results = pd.read_table(islandpath_output, 
                                       sep='\t')
#Put Isolate_name (the name of the isolate) into a column in the table
    Isolate_islandpath_results['Isolate_Pathname'] = isolate_name
#Append the table to your empty dataframe
    all_islandpath_results.append(Isolate_islandpath_results)
#concatentate all the tables together
all_islandpath_results = pd.concat(all_islandpath_results)
#Print to see new table
print(all_islandpath_results)
```
