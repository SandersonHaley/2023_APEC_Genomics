```python
#Import mobsuite results from all the HLF genomes and combine into one big table.
import pandas as pd
from pathlib import Path
#create an empty dataframe
all_VFDB_results = []
#Determine the path that will lead to your mobsuite results files for all HLF genomes
VFDB_results_directory = "/media/haley/My Book/Ecoli_results/Ecoli_ARETE_Results_Organized/"
for VFDB_output in Path(VFDB_results_directory).glob('EC_*/*_VFDB.txt'):
    
    isolate_name = VFDB_output.parts[6]

    print(f"Parsing {VFDB_output}")
#Read tables in mobsuite output files with new header we designate
    Isolate_VFDB_results = pd.read_table(VFDB_output, 
                                       sep='\t',  
                                       names=['qseqid', 'sseqid', 'pident', 'slen', 'qlen', 'length', 'mismatch', 'gapopen', 'qstart', 'qend', 'sstart', 'send', 'evalue', 'bitscore', 'sequence'], 
                                            usecols=['qseqid', 'sseqid', 'pident', 'slen', 'qlen', 'length', 'mismatch', 'gapopen', 'qstart', 'qend', 'sstart', 'send', 'evalue', 'bitscore', 'sequence'])
#Put Plasmid_name (the name of the isolate) into a column in the table
    Isolate_VFDB_results['Isolate_Pathname'] = isolate_name
#Append the table to your empty dataframe
    all_VFDB_results.append(Isolate_VFDB_results)
#concatentate all the tables together
all_VFDB_results = pd.concat(all_VFDB_results)
#Print to see new table
print(all_VFDB_results)
```

```python
VFDB_reports_indexed=all_VFDB_results.set_index('Isolate_Pathname')
```

```python
#setting 80% ID cutoff
VFDB_80_cutoff = VFDB_reports_indexed.loc[VFDB_reports_indexed['pident'] > 80]
VFDB_80_cutoff
```
