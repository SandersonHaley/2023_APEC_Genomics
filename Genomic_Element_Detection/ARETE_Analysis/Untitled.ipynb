```python
import os
import pandas as pd
```

```python
mobsuite = pd.read_csv("/media/haley/My Book1/Ecoli_results/Ecoli_ARETE_Results_Organized/02072023_EcoliARETE_isolates_crosstab_nonovel_mobsuite.csv")
islandcompare = pd.read_csv("/media/haley/My Book1/Ecoli_results/Ecoli_ARETE_Results_Organized/IslandCompare/07052023_EcoliARETE_islandcompare_crosstab.csv")
AMR = pd.read_csv("/media/haley/My Book1/Ecoli_results/Ecoli_ARETE_Results_Organized/02092023_EcoliARETE_RGI_strictperfectfiltered_crosstab.csv")
VFDB = pd.read_csv("/media/haley/My Book1/Ecoli_results/Ecoli_ARETE_Results_Organized/02072023_EcoliARETE_VFDB_filtered80_crosstab_gene_name.csv")
BACMETs = pd.read_csv("/media/haley/My Book1/Ecoli_results/Ecoli_ARETE_Results_Organized/VIDO_ARETE/06302023_EcoliARETE_BACMET_crosstab_gene_name.csv")
```

```python
mobsuite = mobsuite.drop("All", axis=1)
```

```python
mobsuite = mobsuite[mobsuite['row_0'] != "All"]
mobsuite = mobsuite.set_index("row_0")
```

```python
islandcompare = islandcompare.set_index("0")
```

```python
AMR = AMR.drop("All", axis=1)
AMR = AMR[AMR['row_0'] != "All"]
AMR = AMR.set_index("row_0")
AMR
```

```python
VFDB = VFDB.drop("All", axis=1)
VFDB = VFDB[VFDB['Isolate_Pathname'] != "All"]
VFDB = VFDB.set_index("Isolate_Pathname")
VFDB
```

```python
BACMETs = BACMETs.drop("All", axis=1)
BACMETs = BACMETs[BACMETs['row_0'] != "All"]
BACMETs = BACMETs.set_index("row_0")
BACMETs
```
