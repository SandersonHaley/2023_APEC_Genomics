```python
import os
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf #https://anaconda.org/anaconda/statsmodels
import statsmodels.api as sm #https://scikit-posthocs.readthedocs.io/en/latest/installation/
import scikit_posthocs as sp
import itertools as it
from statannotations.Annotator import Annotator
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import mobsuite results from all the HLF genomes and combine into one big table.
import pandas as pd
from pathlib import Path
#create an empty dataframe
all_diamond_results = []
#Determine the path that will lead to your mobsuite results files for all HLF genomes
diamond_results_directory = "/media/haley/My Book/Ecoli_results/Ecoli_ARETE_Results_Organized/Rosemary_Colicin/"
for diamond_output in Path(diamond_results_directory).glob('Genome_Files_Setup_Colicin/EC*_colicin_matches.tsv'):

    Plasmid_name = diamond_output.parts[-1]

    print(f"Parsing {diamond_output}")
#Read tables in mobsuite output files with new header we designate
    Plasmid_diamond_results = pd.read_table(diamond_output, 
                                       sep='\t', 
                                       header=0, 
                                       names=['File', 'qseqid', 'sseqid', 'pident', 'slen', 'qlen', 'length', 'mismatch', 'gapopen', 'qstart', 'qend', 'sstart', 'send', 'evalue', 'bitscore'], 
                                            usecols=['File', 'qseqid', 'sseqid', 'pident', 'slen', 'qlen', 'length', 'mismatch', 'gapopen', 'qstart', 'qend', 'sstart', 'send', 'evalue', 'bitscore'])
#Put Plasmid_name (the name of the isolate) into a column in the table
    Plasmid_diamond_results['Plasmid_Pathname'] = Plasmid_name
#Append the table to your empty dataframe
    all_diamond_results.append(Plasmid_diamond_results)
#concatentate all the tables together
all_diamond_results = pd.concat(all_diamond_results)
#Print to see new table
print(all_diamond_results)
```

```python
colicins=all_diamond_results.set_index("Plasmid_Pathname")
```
