```python
import pandas as pd
from pathlib import Path
all_quast_results = []
quast_results_directory = "/home/haley/Desktop/Complete_Polished_Genomes/Ecoli/"
for quast_output in Path(quast_results_directory).glob('*_quast_output/transposed_report.tsv'):




    isolate_name = quast_output.parts[6]

    print(f"Parsing {quast_output}")
    isolate_quast_results = pd.read_csv(quast_output, sep='\t')
    isolate_quast_results['Isolate_Pathname'] = isolate_name

    all_quast_results.append(isolate_quast_results)

all_quast_results = pd.concat(all_quast_results)

print(all_quast_results)
```

```python
all_quast_results.to_csv("Ecoli_first_86_polished_genomes_quast.csv")
```
