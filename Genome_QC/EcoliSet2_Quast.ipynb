```python
#find all the quast results for HLF genomes and aggregate into one dataframe
import pandas as pd
from pathlib import Path
all_quast_results_set2 = []

quast_results_directory = "/media/haley/My Book1/Ecoli_and_Other_WGS/Complete_Polished_Genomes/EcoliSet2_Finished_Genomes/EcoliSet2_Finished_Genomes/"
for quast_output in Path(quast_results_directory).glob("*_quast_output/transposed_report.tsv"):

    isolate_name = quast_output.parts[8]

    print(f"Parsing {quast_output}")
    isolate_quast_results = pd.read_csv(quast_output, sep='\t')
    isolate_quast_results['Isolate_Pathname'] = isolate_name

    all_quast_results_set2.append(isolate_quast_results)

all_quast_results_set2 = pd.concat(all_quast_results_set2)

print(all_quast_results_set2)
```

```python
all_quast_results_set2.to_csv("Set2_Quast_results.csv")
```
