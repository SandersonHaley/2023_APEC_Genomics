```python
import pandas as pd
from pathlib import Path
all_nanostat_results1 = []

nanostat_results_directory = "/media/haley/My Book/Ecoli_Hybrid_Assembly_Set2/Ecoli_second_set/Ecoli_Set_Two_fastqs/"

for nanostat_output in Path(nanostat_results_directory).glob('*/nanostat_results_rawreads/*_nanostat_results'):

    isolate_name = nanostat_output.parts[7]

    print(f"Parsing {nanostat_output}")

    isolate_nanostat_results = pd.read_csv(nanostat_output, sep=':')
    isolate_nanostat_results_t = isolate_nanostat_results.transpose()
    isolate_nanostat_results_t['Isolate_Pathname'] = isolate_name
    isolate_nanostat_results_t_r2 = isolate_nanostat_results_t.iloc[[1]]
    
    all_nanostat_results1.append(isolate_nanostat_results_t_r2)

all_nanostat_results = pd.concat(all_nanostat_results1)


#all_quast_results_CN =['General summary', 'Mean read length', 'Mean read quality', 'Median read length', 'Median read quality', 'Number of reads', 'Read length N50','STDEV read length', 'Total bases', 'Number, percentage and megabases of reads above quality cutoffs', '>Q5', '>Q7', '>Q10', '>Q12', '>Q15', 'Top 5 highest mean basecall quality scores and their read lengths', '1', '2', '3', '4', '5', 'Top 5 longest reads and their mean basecall quality score', '1', '2', '3', '4', '5', 'Isolate_Name']

print(all_nanostat_results)
```
